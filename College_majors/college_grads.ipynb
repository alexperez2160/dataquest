```python
import pandas as pd 
import matplotlib as plt

%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
recent_grads.head()
recent_grads.describe()
```

```python
raw_data_count = recent_grads.count()[0]
raw_data_count
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.count()[0]
cleaned_data_count
```

```python
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Employed vs. Sample_size')
```

```python
ax2 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax2.set_title('Median vs. Sample_size')
```

```python
ax3 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax3.set_title('Unemployment_rate vs. Sample_size')
```

```python
ax4 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax4.set_title('Median vs. Full_time')
```

```python
ax5 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax5.set_title('Unemployment_rate vs. ShareWomen')
```

```python
ax6 = recent_grads.plot(x='Men', y='Women', kind='scatter')
ax6.set_title('Women vs. Men')
```

```python
ax7 = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax7.set_title('Median vs. Women')
```

There is a positive correlation between sample size and median income. 

There doesn't seem to a correlation between subjects majority female and income. 

There doesn't seem to be a link between the number of full-time employees and median salaries.

```python
major_by_median = recent_grads.sort_values(by=['Median'], ascending = False)

ax = major_by_median[:10]['ShareWomen'].plot(kind='bar')
ax.set_xticklabels(major_by_median.Major)
```

```python
major_by_median_tail = recent_grads.sort_values(by=['Median'], ascending = False).tail(10)

ax2 = major_by_median_tail[:10]['ShareWomen'].plot(kind='bar')
ax2.set_xticklabels(major_by_median_tail.Major)
```

Some of the highest paid majors have very low numbers of women and some of the lowest paying majors have a high % of women.
